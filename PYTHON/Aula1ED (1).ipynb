Comando de Entrada de Dados

```python
#Esse comando para entrada de dados pelo teclado
nome = input("Nome : ")
idade = int(input("idade : "))
print(f"Meu nome e {nome} e minha idade e {idade}")
```

```python
#printa a idade multiplicada por 2
nome  = input("Digite seu nome : ")
idade = int(input("Digite sua idade : ")) #casting
print(f"Meu nome e {nome} e minha idade e {idade*2}")
```

```python
#Programa para calcular media e replicar linhas shift alt seta pra baixo

nome = input('Nome: ')
materia = input("Materia: ")
n1 = float(input("Digite a nota 1: "))
n2 = float(input("Digite a nota 2: "))
n3 = float(input("Digite a nota 3: "))
media = (n1 + n2 + n3) / 3
print(f"O aluno {nome} da materia {materia}, tirou a nota {media:.1f}")
```

```python
#Segunda versão do primeiro codigo
nome = input('Nome: ')
materia = input("Materia: ")
n1 = float(input("Digite a nota 1: "))
n2 = float(input("Digite a nota 2: "))
n3 = float(input("Digite a nota 3: "))
print(f"O aluno {nome} da materia {materia}, tirou a nota {(n1 + n2 + n3) / 3 :.1f}")
```

Estrutura de Decisão


```python
#Programa que demonstra IF e ELSE

a = 10

if a > 10 :
  print(f"E claro que e maior")
elif a == 10:
  print(f"é igual")
else :
  print(f"e menor")
```
