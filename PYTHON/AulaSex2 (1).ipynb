**Listas**

**Em python, as listas são um tipo de estrutura de dados que permite armazenar uma coleção de itens em uma unica variável. As listas são mutáveis, o que significa que podem ser modificadas após serem criadas.**

```python
lista = [1, 2, 3, 4, 5]
```

```python
type(lista)
```

```python
lista[3]
```

```python
lista[6] #Erro por estourar o tamanho da lista
```

**Metodo = APPEND() - insere um elemento sempre no final da lista**

```python
lista.append(7)
lista
```

**Metodo INSERT() - insere um elemento na posição desejada**

```python
lista.insert(0, 25)
lista
```

**Metdo DEL() - deleta um item na lista baseado na posição desejada**

```python
del lista[5]
lista
```

```python
del lista #apaga a lista inteira
```

**Metodo REMOVE() - remove um item baseado no valor e não na posição**

```python
numero = [2, 4, 6, 8, 9, 10]
```

```python
numero.remove(6) #remove o 6 da lista
numero
```

**Metodo CLEAR() - apaga o conteúdo da lista inteira retornando uma lista vazia**

```python
numero.clear() #apaga os elementos da lista, não a estrutura
numero
```

**Metodo POP() - elimina o ultimo elemento  inserido e mostra ele**

```python
numeros = [1, 5, 7, 90, 50, 70]
```

```python
numeros.pop() #sem parametro, remove e mostra o ultimo elemento inserido
numeros
```

```python
numeros.pop(2) #com parametro remove o elemento na posição desejada
```

**3. Faça um programa que leia 20 numeros inteiros e armazene em uma lkista chamada pares, os numero pares em uma lista e os numero impares em outra chamada impares, no final imprima as duas listas**

```python
Pares = []
Impares = []
```

```python
for i in range(20):
  if i % 2 == 0  :
   Pares.append(i)
  elif i % 2 == 1 :
    Impares.append(i)

print(Pares)
print(Impares)
```

```python
Pares.clear()
Impares.clear()
```
